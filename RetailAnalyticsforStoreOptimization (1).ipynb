## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Standardize column names
for df in [df_stores, df_sales, df_features]:
    df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
````

````python
#  Handle missing values ---
# Fill missing markdowns with 0
markdown_cols = [col for col in df_features.columns if "markdown" in col.lower()]
df_features[markdown_cols] = df_features[markdown_cols].fillna(0)

# Forward fill for CPI, Unemployment
for col in ["cpi", "unemployment"]:
    if col in df_features.columns:
        df_features[col] = df_features[col].fillna(method="ffill")
````

````python
# Double-check remaining missing values
print("🔍 Remaining missing values in Features dataset:")
print(df_features.isnull().sum())
````

````python
# Merge datasets
df_merged = pd.merge(df_sales, df_features, on=["store","date","isholiday"], how="left")
df_merged = pd.merge(df_merged, df_stores, on="store", how="left")

````

````python
# Ensure correct dtypes (with dayfirst=True)
df_merged["date"] = pd.to_datetime(df_merged["date"], dayfirst=True, errors='coerce')
df_merged["isholiday"] = df_merged["isholiday"].astype(int)

# Check if any rows failed to parse
print("Unparsed dates:", df_merged["date"].isna().sum())
````

````python
# - Final sanity checks ---
print("✅ Final merged dataset shape:", df_merged.shape)
print("\n📊 Columns in merged dataset:\n", df_merged.columns.tolist())
````

````python
#  Ensure correct data types ---

df_merged["store"] = df_merged["store"].astype("category")
df_merged["dept"] = df_merged["dept"].astype("category")
````

````python
# ---  Quick check ---
print("\nFirst 5 rows of final dataset:")
print(df_merged.head())
````

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ensure datetime format
df_merged['date'] = pd.to_datetime(df_merged['date'], errors='coerce')

# Extract month & year
df_merged['month'] = df_merged['date'].dt.month
df_merged['year'] = df_merged['date'].dt.year

# Group average sales by year & month
heatmap_data = df_merged.groupby(['year','month'])['weekly_sales'].mean().reset_index()

# Pivot for heatmap
heatmap_pivot = heatmap_data.pivot(index='year', columns='month', values='weekly_sales')

# Plot heatmap
plt.figure(figsize=(12,6))
sns.heatmap(heatmap_pivot, cmap="YlGnBu", annot=True, fmt=".0f")
plt.title("Heatmap of Average Monthly Sales by Year")
plt.xlabel("Month")
plt.ylabel("Year")
plt.show()
````

##### 1. Why did you pick the specific chart?

A heatmap is ideal for spotting seasonality and year-over-year trends in one view.

It shows which months consistently perform better (holidays) and which months are weaker (off-season).

It’s more compact than multiple line plots and gives a clear “calendar-style” overview.

**Outlier Detection & Treatment**

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# -------------------------------
# 1. Visualize outliers
# -------------------------------
plt.figure(figsize=(10,6))
sns.boxplot(y=df_merged['weekly_sales'])
plt.title("Boxplot of Weekly Sales (with outliers)")
plt.show()

# -------------------------------
# 2. Z-score method
# -------------------------------
z_scores = np.abs(stats.zscore(df_merged['weekly_sales']))
outliers_z = df_merged[z_scores > 3]
print("Outliers detected (Z-score > 3):", len(outliers_z))

# -------------------------------
# 3. IQR method
# -------------------------------
Q1 = df_merged['weekly_sales'].quantile(0.25)
Q3 = df_merged['weekly_sales'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_iqr = df_merged[(df_merged['weekly_sales'] < lower_bound) |
                         (df_merged['weekly_sales'] > upper_bound)]
print("Outliers detected (IQR method):", len(outliers_iqr))

# -------------------------------
# 4. Treatment (Capping / Winsorization)
# -------------------------------
# We cap extreme values at 1st and 99th percentile
lower_cap = df_merged['weekly_sales'].quantile(0.01)
upper_cap = df_merged['weekly_sales'].quantile(0.99)

df_merged['weekly_sales_capped'] = np.where(
    df_merged['weekly_sales'] < lower_cap, lower_cap,
    np.where(df_merged['weekly_sales'] > upper_cap, upper_cap, df_merged['weekly_sales'])
)

# -------------------------------
# 5. Compare distributions before/after treatment
# -------------------------------
plt.figure(figsize=(12,6))
sns.kdeplot(df_merged['weekly_sales'], label="Original", shade=True)
sns.kdeplot(df_merged['weekly_sales_capped'], label="Capped", shade=True)
plt.title("Weekly Sales Distribution: Before vs After Outlier Treatment")
plt.legend()
plt.show()
````

What all outlier treatment techniques have you used and why did you use those techniques?
Outlier Treatment Documentation

Detection Techniques Used
Boxplot Visualization

To quickly see the spread of weekly sales and detect extreme values.

Useful for identifying spikes like Black Friday sales.

Z-Score Method

Flags values beyond ±3 standard deviations from the mean.

Good for detecting extreme outliers in normally distributed data.

IQR Method (Interquartile Range)

Based on the 1.5 × IQR rule (values below Q1–1.5×IQR or above Q3+1.5×IQR are outliers).

More robust for skewed data like sales.

Treatment Technique Used
Winsorization (Capping)

Replaced extreme values in weekly_sales below the 1st percentile with the 1st percentile value, and values above the 99th percentile with the 99th percentile value.

Created a new column: weekly_sales_capped for analysis and hypothesis testing.

Why These Techniques Were Used
Why Not Remove Outliers?

Why Winsorization (Capping)?

Keeps the data realistic (holiday peaks still visible).

Prevents extreme values from dominating averages, variances, and p-values in hypothesis testing.

Maintains dataset size (no loss of rows).



## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Hypothesis 1 — Holiday Impact

H₀ (Null Hypothesis): There is no significant difference in weekly sales between holiday weeks and non-holiday weeks.

H₁ (Alternative Hypothesis): Weekly sales during holiday weeks are significantly higher than non-holiday weeks.Answer Here.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
import scipy.stats as stats

# =========================
# 1. Holiday Impact (t-test)
# =========================
holiday_sales = df_merged[df_merged['isholiday']==True]['weekly_sales_capped']
nonholiday_sales = df_merged[df_merged['isholiday']==False]['weekly_sales_capped']

t_stat, p_val = stats.ttest_ind(holiday_sales, nonholiday_sales, equal_var=False)
print("Holiday vs Non-Holiday Sales Test:")
print("t-statistic:", t_stat, "p-value:", p_val)
if p_val < 0.05:
    print("Conclusion: Reject H₀ → Holiday weeks have significantly different sales.")
else:
    print("Conclusion: Fail to reject H₀ → No significant difference found.")
````

##### Which statistical test have you done to obtain P-Value?

I used a Two-Sample Independent t-test (Welch’s t-test, since variances may differ).

##### Why did you choose the specific statistical test?

We are comparing the means of two independent groups (holiday weeks vs. non-holiday weeks).

The t-test is appropriate for testing if the difference in average sales between these two groups is statistically significant.

Welch’s version was used because the variance of sales may not be equal across groups.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Hypothesis 2 — Store Type Effect

H₀ (Null Hypothesis): Average weekly sales are the same across all store types (A, B, C).

H₁ (Alternative Hypothesis): At least one store type has different average weekly sales compared to others.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
#2. Store Type Effect (ANOVA)
# =========================
storeA = df_merged[df_merged['type']=='A']['weekly_sales_capped']
storeB = df_merged[df_merged['type']=='B']['weekly_sales_capped']
storeC = df_merged[df_merged['type']=='C']['weekly_sales_capped']

f_stat, p_val = stats.f_oneway(storeA, storeB, storeC)
print("\nStore Type Effect Test:")
print("F-statistic:", f_stat, "p-value:", p_val)
if p_val < 0.05:
    print("Conclusion: Reject H₀ → At least one store type has different average sales.")
else:
    print("Conclusion: Fail to reject H₀ → No significant difference between store types.")
````

##### Which statistical test have you done to obtain P-Value?

I used a One-Way ANOVA (Analysis of Variance) test

##### Why did you choose the specific statistical test?

We are comparing the means of more than two groups (Store Types A, B, and C).

ANOVA checks whether at least one store type has a significantly different mean weekly sales compared to others.

If p-value < 0.05, it suggests not all store types are equal → further post-hoc tests (like Tukey’s HSD) can pinpoint where the difference lies.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Hypothesis 3 — Unemployment Impact

H₀ (Null Hypothesis): There is no correlation between unemployment rate and weekly sales.

H₁ (Alternative Hypothesis): There is a negative correlation between unemployment rate and weekly sales.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# =========================
# 3. Unemployment Impact (Correlation)
# =========================
corr, p_val = stats.pearsonr(df_merged['unemployment'], df_merged['weekly_sales_capped'])
print("\nUnemployment vs Sales Correlation Test:")
print("Correlation coefficient:", corr, "p-value:", p_val)
if p_val < 0.05:
    if corr < 0:
        print("Conclusion: Reject H₀ → Significant negative correlation (higher unemployment → lower sales).")
    else:
        print("Conclusion: Reject H₀ → Significant positive correlation.")
else:
    print("Conclusion: Fail to reject H₀ → No significant correlation found.")
````

##### Which statistical test have you done to obtain P-Value?

I used the Pearson Correlation Test.

##### Why did you choose the specific statistical test?

We are testing whether there is a linear relationship between two continuous variables: unemployment rate and weekly sales.

Pearson correlation provides both:

Correlation coefficient (r) → strength & direction of relationship.

p-value → significance of that relationship.

It is appropriate here because both variables are continuous and approximately linear in relationship.

## ***6. Feature Engineering & Data Pre-processing***

````python
#Train-Test Split (chronological to avoid leakage)
# =====================================================
train = df_merged[df_merged['date'] < "2012-01-01"].copy()
test = df_merged[df_merged['date'] >= "2012-01-01"].copy()

print("Train shape:", train.shape)
print("Test shape:", test.shape)
````

````python
#6. Target Encoding (Dept) — from Train Only
# =====================================================
# Step 1: Compute department-wise average sales
dept_avg_sales = train.groupby('dept', observed=False)['weekly_sales_capped'].mean().to_dict()

# Step 2: Compute global mean as fallback
global_mean = train['weekly_sales_capped'].mean()

# Step 3: Map dept averages to train set (convert to float to avoid categorical issues)
train['dept_encoded'] = train['dept'].map(dept_avg_sales).astype(float)

# Step 4: Map dept averages to test set, fill unseen departments with global mean
test['dept_encoded'] = test['dept'].map(dept_avg_sales).astype(float).fillna(global_mean)
````

````python
# 7. Feature Engineering (Leakage-Free)
# =====================================================
def create_features(df):
    df = df.sort_values(by=['store','dept','date'])

    # Date Features
    df['year'] = df['date'].dt.year
    df['month'] = df['date'].dt.month
    df['week'] = df['date'].dt.isocalendar().week
    df['dayofweek'] = df['date'].dt.dayofweek
    df['year_month'] = df['date'].dt.to_period('M')

    # Promotion Features
    df['total_markdown'] = df[markdown_cols].sum(axis=1)

    # Lag & Rolling Features (shift prevents leakage)
    df['lag_1'] = df.groupby(['store','dept'])['weekly_sales_capped'].shift(1)
    df['lag_4'] = df.groupby(['store','dept'])['weekly_sales_capped'].shift(4)
    df['rolling_4w'] = df.groupby(['store','dept'])['weekly_sales_capped'].transform(lambda x: x.shift(1).rolling(4).mean())
    df['rolling_12w'] = df.groupby(['store','dept'])['weekly_sales_capped'].transform(lambda x: x.shift(1).rolling(12).mean())

    # Interaction Features
    df['promo_holiday'] = df['total_markdown'] * df['isholiday']
    df['holiday_lag1'] = df['isholiday'].shift(1)

    # Normalized Features
    df['sales_per_size'] = df['weekly_sales_capped'] / df['size']

    return df

train = create_features(train)
test = create_features(test)
````

````python
 #8. Final Dataset Preparation
# =====================================================
drop_cols = ['weekly_sales','weekly_sales_capped','date','dept','year_month']
X_train = train.drop(columns=drop_cols)
y_train = train['weekly_sales_capped']

X_test = test.drop(columns=drop_cols)
y_test = test['weekly_sales_capped']
````
